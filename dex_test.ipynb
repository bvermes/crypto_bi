```python

CHAIN = "ethereum"
TOP_PAIRS_URL = f"https://api.dexscreener.com/latest/dex/trading-pairs?chain={CHAIN}"

try:
    response = requests.get(TOP_PAIRS_URL)
    response.raise_for_status()
    data = response.json()
except Exception as e:
    print("Error fetching top pairs:", e)
    data = {}

# Assume the returned JSON has a key "pairs" which is a list of trading pair objects.
all_pairs = data.get("pairs", [])

if not all_pairs:
    print("No trading pairs found.")
else:
    # For demonstration, we assume the list is already sorted (e.g., by volume) and take the top 10.
    top_pairs = all_pairs[:10]
    print(f"Found {len(top_pairs)} top trading pairs on chain '{CHAIN}':")
    for i, pair in enumerate(top_pairs, 1):
        # Each pair object is assumed to have 'pair' (name) and 'id' fields.
        print(f"{i}. {pair.get('pair', 'N/A')} (id: {pair.get('id', 'N/A')})")
```
